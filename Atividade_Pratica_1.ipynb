## Escreva um programa para transformar uma imagem de níveis de cinza na qual cada pixel é representado por 8 bits (256 níveis de cinza), conforme as seguintes operações:

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from IPython.display import Image
```

### a) inverter os valores de intensidade da imagem, tal que o valor 255 passa a ser 0, 254 passa a ser 1, assim por diante;

```python
img1 = cv2.imread("nickel.jpg", cv2.IMREAD_GRAYSCALE);
```

```python
imagem_invertida = cv2.bitwise_not(img1)
```

```python
res_A = np.hstack((img1,imagem_invertida))

#cv2.imshow('res_A.jpg',res_A)
#cv2.waitKey(0)
#cv2.destroyAllWindows()

cv2.imwrite("res_A.jpg",res_A)
Image(filename='res_A.jpg') 
```

-------------------------------------------------------------------------------------------------------------------------------

### b) os valores de intensidade da imagem presentes nas colunas pares são trocados com os valores de intensidade das colunas ímpares;

https://docs.opencv.org/3.0-beta/doc/py_tutorials/py_core/py_basic_ops/py_basic_ops.html

```python
img2 = cv2.imread("nickel.jpg", cv2.IMREAD_GRAYSCALE);
```

```python
rows,cols = img2.shape
```

```python
print("shape: ", img2.shape)
print("size: ", img2.size)
print("dtype: ", img2.dtype)
```

```python
print(np.matrix(img2))
```

```python
aux = [[0] * cols] * rows
```

```python
for i in range(rows-1):
    for j in range(cols-1):
        if(j % 2 == 0):
            aux[i][j] = img2[i][j]
            img2[i][j] = img2[i][j-1]
            img2[i][j-1] = aux[i][j]
```

```python
#cv2.imshow('cols-even-odd.png',img2)
#cv2.waitKey(0)
#cv2.destroyAllWindows()

cv2.imwrite("img2.jpg",img2)
Image(filename='img2.jpg') 
```

-------------------------------------------------------------------------------------------------------------------------------

### c) os valores de intensidade da imagem presentes nas linhas pares são trocados com os valores de intensidade das linhas ímpares;

```python
img3 = cv2.imread("nickel.jpg", cv2.IMREAD_GRAYSCALE);
```

```python
rows,cols = img3.shape
```

```python
print("shape: ", img3.shape)
print("size: ", img3.size)
print("dtype: ", img3.dtype)
```

```python
print(np.matrix(img3))
```

```python
aux = [[0] * cols] * rows
```

```python
for i in range(rows-1):
    for j in range(cols-1):
        if(i % 2 == 0):
            aux[i][j] = img3[i][j]
            img3[i][j] = img3[i-1][j]
            img3[i-1][j] = aux[i][j]
```

```python
#cv2.imshow('img3.jpg',img3)
#cv2.waitKey(0)
#cv2.destroyAllWindows()

cv2.imwrite("img3.jpg",img3)
Image(filename='img3.jpg') 
```

```python
#SHOW 
res1 = np.hstack((img2,img3)) #stacking images side-by-side

#cv2.imshow('res1.png',res1)
#cv2.waitKey(0)
#cv2.destroyAllWindows()

cv2.imwrite("res1.jpg",res1)
Image(filename='res1.jpg') 
```

### d) realizar o processo de alargamento de contraste (histogram strechting).

https://docs.opencv.org/master/d5/daf/tutorial_py_histogram_equalization.html

```python
img4 = cv2.imread('nickel.jpg',0)
plt.hist(img4.ravel(),256,[0,256]); 
plt.show()
```

```python
img_D = cv2.imread('nickel.jpg',0)
equ = cv2.equalizeHist(img_D)
res2 = np.hstack((img_D,equ)) #stacking images side-by-side

#cv2.imshow('res2.png',res2)
#cv2.waitKey(0)
#cv2.destroyAllWindows()

cv2.imwrite("res2.jpg",res2)
Image(filename='res2.jpg')
```
