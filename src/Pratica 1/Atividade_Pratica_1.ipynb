## Escreva um programa para transformar uma imagem de níveis de cinza na qual cada pixel é representado por 8 bits (256 níveis de cinza), conforme as seguintes operações:

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from IPython.display import Image
```

### a) inverter os valores de intensidade da imagem, tal que o valor 255 passa a ser 0, 254 passa a ser 1, assim por diante;

### ---------Utilizando função---------

```python
img_nickel_with_function = cv2.imread("nickel.jpg", cv2.IMREAD_GRAYSCALE);
```

```python
img_nickel_with_function_inverted = cv2.bitwise_not(img_nickel_with_function)
```

```python
solution_with_function = np.hstack((img_nickel_with_function,img_nickel_with_function_inverted))
cv2.imwrite("solution_with_function.jpg",solution_with_function)
Image(filename='solution_with_function.jpg') 
```

### ---------Sem utilizar função---------

```python
img_nickel_without_function = cv2.imread("nickel.jpg", cv2.IMREAD_GRAYSCALE);
```

```python
print(np.matrix(img_nickel_without_function))
```

```python
rows,cols = img_nickel_without_function.shape
```

```python
img_nickel_without_function_inverted = [[0] * cols] * rows
```

```python
for i in range(rows-1):
    for j in range(cols-1):
        img_nickel_without_function_inverted[i][j] =  abs(img_nickel_without_function[i][j] - 255)
        print("antes: ", abs(img_nickel_without_function[i][j] - 255), " invertido: ", img_nickel_without_function[i][j])
        
```

```python
solution_without_function = np.hstack((img_nickel_without_function,img_nickel_without_function_inverted))
cv2.imwrite("solution_without_function.jpg",solution_without_function)
Image(filename='solution_without_function.jpg') 
```

### b) os valores de intensidade da imagem presentes nas colunas pares são trocados com os valores de intensidade das colunas ímpares;

https://docs.opencv.org/3.0-beta/doc/py_tutorials/py_core/py_basic_ops/py_basic_ops.html

```python
img_nickel_cols = cv2.imread("nickel.jpg", cv2.IMREAD_GRAYSCALE);
```

```python
rows,cols = img_nickel_cols.shape
```

```python
print("shape: ", img_nickel_cols.shape)
print("size: ", img_nickel_cols.size)
print("dtype: ", img_nickel_cols.dtype)
```

```python
print(np.matrix(img_nickel_cols))
```

```python
aux = [[0] * cols] * rows
```

```python
for i in range(rows-1):
    for j in range(cols-1):
        if(j % 2 == 0):
            aux[i][j] = img_nickel_cols[i][j]
            img_nickel_cols[i][j] = img_nickel_cols[i][j-1]
            img_nickel_cols[i][j-1] = aux[i][j]
```

```python
cv2.imwrite("img_nickel_cols.jpg",img_nickel_cols)
Image(filename='img_nickel_cols.jpg') 
```

### c) os valores de intensidade da imagem presentes nas linhas pares são trocados com os valores de intensidade das linhas ímpares;

```python
img_nickel_rows = cv2.imread("nickel.jpg", cv2.IMREAD_GRAYSCALE);
```

```python
print("shape: ", img_nickel_rows.shape)
print("size: ", img_nickel_rows.size)
print("dtype: ", img_nickel_rows.dtype)
```

```python
print(np.matrix(img_nickel_rows))
```

```python
aux = [[0] * cols] * rows
```

```python
for i in range(rows-1):
    for j in range(cols-1):
        if(i % 2 == 0):
            aux[i][j] = img_nickel_rows[i][j]
            img_nickel_rows[i][j] = img_nickel_rows[i-1][j]
            img_nickel_rows[i-1][j] = aux[i][j]
```

```python
cv2.imwrite("img_nickel_rows.jpg",img_nickel_rows)
Image(filename='img_nickel_rows.jpg') 
```

```python
#SHOW 
solution_cols_rows = np.hstack((img_nickel_cols,img_nickel_rows)) #stacking images side-by-side
cv2.imwrite("solution_cols_rows.jpg",solution_cols_rows)
Image(filename='solution_cols_rows.jpg') 
```

```python
img_nickel_hs_nf = cv2.imread('nickel.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
# To ascertain total numbers of rows and columns of the image, size of the image
m, n = img_nickel_hs_nf.shape
```

```python
# Transformation to obtain stretching
constant = (255-0)/(img_nickel_hs_nf.max()-img_nickel_hs_nf.min())
img_nickel_hs_nf_stretch = img_nickel_hs_nf * constant
r2, count2 = hist_plot(img_nickel_hs_nf_stretch)
```
