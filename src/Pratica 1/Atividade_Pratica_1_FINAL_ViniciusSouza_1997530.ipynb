### a) inverter os valores de intensidade da imagem, tal que o valor 255 passa a ser 0, 254 passa a ser 1, assim por diante;

#### Lendo a imagem em escala de cinza

```python
img_nickel_a = cv2.imread("nickel.jpg", cv2.IMREAD_GRAYSCALE);
```

```python
print(np.matrix(img_nickel_a))
```

#### Pegando os numeros de linhas e colunas da imagem

```python
rows_a,cols_a = img_nickel_a.shape
```

#### Inicializando a matriz que receberá a imagem invertida

```python
img_nickel_a_inverted = [[0] * cols_a] * rows_a
```

#### Invertendo o valor de cada pixel , subtraindo 255 e pegando seu valor absoluto

```python
for i in range(rows_a-1):
    for j in range(cols_a-1):
        img_nickel_a_inverted[i][j] =  abs(img_nickel_a[i][j] - 255)
        print("antes: ", abs(img_nickel_a[i][j] - 255), " invertido: ", img_nickel_a[i][j])
        
```

#### Fazendo a comparação da imagem original e da imagem invertida

```python
solution_a = np.hstack((img_nickel_a,img_nickel_a_inverted))
cv2.imwrite("solution_a.jpg",solution_a)
Image(filename='solution_a.jpg') 
```

### b) os valores de intensidade da imagem presentes nas colunas pares são trocados com os valores de intensidade das colunas ímpares;

#### Lendo a imagem

```python
img_nickel_cols = cv2.imread("nickel.jpg", cv2.IMREAD_GRAYSCALE);
```

#### Pegando os numeros de linhas e colunas da imagem

```python
rows,cols = img_nickel_cols.shape
```

```python
print("shape: ", img_nickel_cols.shape)
print("size: ", img_nickel_cols.size)
print("dtype: ", img_nickel_cols.dtype)
```

```python
print(np.matrix(img_nickel_cols))
```

#### Inicializando a matriz auxiliar

```python
aux = [[0] * cols] * rows
```

#### Função que percore as colunas da imagem, se a coluna for valor PAR, ela troca seus valores com a coluna ÍMPAR anterior

```python
for i in range(rows-1):
    for j in range(cols-1):
        if(j % 2 == 0):
            aux[i][j] = img_nickel_cols[i][j]
            img_nickel_cols[i][j] = img_nickel_cols[i][j-1]
            img_nickel_cols[i][j-1] = aux[i][j]
```

#### Imprime a imagem

```python
cv2.imwrite("img_nickel_cols.jpg",img_nickel_cols)
Image(filename='img_nickel_cols.jpg') 
```

### Histogram Stretching

```python
display(Math(r'g(x,y) = {\frac{f(x,y) - f_{min}}{f_{max} - f_{min}}}*255 = {\frac{f(x,y)}{f_{max}}}*255'))
```

```python
f_xy = img_einstein_hs
g_xy = ((f_xy - f_xy.min())/(f_xy.max()-f_xy.min()))*255 
```

#### Realizando o calculo do histograma e plotando o gráfico


```python
valor_intensidade_2, contador_intensidade_2 = hist_plot(g_xy)
plt.stem(valor_intensidade_2, contador_intensidade_2)
plt.xlabel('Valor de intensidade')
plt.ylabel('Numero de pixels')
plt.title('Histograma - Imagem stretched')
```
