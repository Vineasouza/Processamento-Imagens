```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from IPython.display import Image
```

-------------------------------------------------------------------------------------------------------------------------------

### Binary Threshold

```python
f = cv2.imread('einstein.jpg',cv2.IMREAD_GRAYSCALE)
x,y = f.shape
```

```python
f_array = f.flatten()
size_array = len(f_array)
T_anterior = 0
T_novo = np.mean(f_array)
```

```python
K = f.max()
R1 = []
R2 = []
μ1 = 0
μ2 = 0



delta = 0.01
while abs(T_anterior - T_novo) > delta:
    T_anterior = T_novo
    for i in range(x):
        for j in range(y):
            if(f[i][j] < T_anterior):
                R1.append(f[i][j])
                
            else:
                R2.append(f[i][j])
    μ1 = np.mean(R1)
    μ2 = np.mean(R2)
    T_novo = np.mean([μ1, μ2])


for i in range(x-1):
    for j in range(y-1):
        if(f[i][j] > T_novo):
            f[i][j] = K
        else:
            f[i][j] = 0
```

```python
cv2.imwrite("f_RC_binary_threshold.jpg",f)
Image(filename='f_RC_binary_threshold.jpg')
```

```python
f_array = f.flatten()
size_array = len(f_array)
T_anterior = 0
T_novo = np.mean(f_array)
```

```python
K = f.max()
R1 = []
R2 = []
μ1 = 0
μ2 = 0

delta = 0.01
while abs(T_anterior - T_novo) > delta:
    T_anterior = T_novo
    for i in range(x):
        for j in range(y):
            if(f[i][j] < T_anterior):
                R1.append(f[i][j])
                
            else:
                R2.append(f[i][j])
    μ1 = np.mean(R1)
    μ2 = np.mean(R2)
    T_novo = np.mean([μ1, μ2])

for i in range(x):
    for j in range(y):
        if(f[i][j] > T_novo):
            f[i][j] = f[i][j]
        else:
            f[i][j] = 0
```
