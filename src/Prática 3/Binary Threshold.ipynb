```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from IPython.display import Image
```

```python
f = cv2.imread('einstein.jpg',cv2.IMREAD_GRAYSCALE)
x,y = f.shape
```

```python
f_array = f.flatten()
```

```python
T = np.mean(f_array)
T
```

```python
K = f.max()
```

```python
for i in range(x):
    for j in range(y):
        if(f[i][j] > T):
            f[i][j] = K
        else:
            f[i][j] = 0
```

```python
cv2.imwrite("f_binary_threshold.jpg",f)
Image(filename='f_binary_threshold.jpg')
```
